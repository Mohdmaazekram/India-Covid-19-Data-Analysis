```python
import pandas as pd      # Importing Pabdas Library.
import numpy as np    # Importing Numpy Library.
import matplotlib.pyplot as plt   # Importing Mamatplotlib.pyplot Library.
import seaborn as sns   # Importing seaborn Library.




d1 = pd.read_csv('covid_19_india.csv')    # Loading Data With The Help of Pandas.
d1
```

```python
d1.replace('Bihar****','Bihar',inplace=True)  # Clubing Data Using Replace.

d1.replace('Dadra and Nagar Haveli','Dadra and Nagar Haveli and Daman and Diu',inplace=True) # Clubing Data Using Replace.

d1.replace('Telangana','Telengana',inplace=True)  # Clubing Data Using Replace.



d1.replace('-',0,inplace=True)   # Filling Null Data Using Replace.



d2 = d1.drop([500,528,617],axis=0)  # Droping Data With Help of Drop Function.

d2 = d2[d2['State/UnionTerritory'] != 'Cases being reassigned to states']  # Droping Some More Data.



d3 = d2.astype({"ConfirmedIndianNational":'int64',"ConfirmedForeignNational":'int64'})  # Converting Object into Int.

d3["Date"] = pd.to_datetime(d3["Date"])  # Converting Object into DateTime.




d3.info()   # Using info to Check All DataType.
d3   # Print d3 for Showing All Data.
```

#### Adding Alive Cases Column:-

```python
d3['Alive'] = d3['Confirmed']-(d3['ConfirmedIndianNational']+d3['ConfirmedForeignNational']+d3['Deaths']+d3['Cured']) # Creating One New Alive Column.
d3
```

#### Adding Year,Month,Day Column :-

```python
d3['Year'] = pd.DatetimeIndex(d3['Date']).year  # Extracting year from Date Collumn.


Date = pd.date_range('01-Jan 2020','01-Jan 2022',freq='M')
d4 = pd.DataFrame({'Date' : Date})
d3['Month'] = d3['Date'].dt.strftime('%b')     # Extracting Month from Date Collumn.


d3['Day'] = pd.DatetimeIndex(d3['Date']).day   # Extracting day from Date Collumn.
d3
```

#### Creating Percetage Column:-

```python
percentage = d3.groupby('State/UnionTerritory')[['Confirmed','Deaths','Cured','Alive']].sum()  # Creating Percentage % Column.
percentage['Confirmed%'] = (percentage['Confirmed']/percentage['Confirmed'].sum())*100
percentage['Deaths_%'] = (percentage['Deaths']/percentage['Deaths'].sum())*100
percentage['Cured_%'] = (percentage['Cured']/percentage['Cured'].sum())*100
percentage['Alive_%'] = (percentage['Alive']/percentage['Alive'].sum())*100
percentage
```

```python
fig = plt.figure(figsize=(20,10))
sns.set_style(style='whitegrid')     # State Wise Deaths_%.
sns.lineplot(x='State/UnionTerritory',y='Deaths_%',data=percentage,linewidth=2.5,color='r')
plt.xticks(rotation=40,horizontalalignment='right',fontsize=20)
plt.title('Percentage Wise Deaths Cases')
plt.tight_layout()
plt.show()
```

#### Adding Daily_Base_Cases Column:-

```python
dt = d3['Confirmed'].tolist()                # Converting Confirmed value into list.
dt1 = []                                          # Created Empty list.
copied_list = dt.copy()                               # Copied list into New Variable.
index_convert = dt[0]                               # Remove index[0] Value.
copied_list.pop(0)                  
copied_list.append(int(0))
for item in range(0,len(dt)):
    diff = copied_list[item]-dt[item]
    dt1.append(diff)
dt1.insert(0,index_convert)                 # inserting covert[0] value in diff list.
dt1.pop()                               # Removing the last Value.
dt1
d3['Daily_Base_Cases'] = dt1       # Adding Column as Daily_Base_Cases.
d3  
```
